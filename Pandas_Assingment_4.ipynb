```python
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```python
def gnd(n):
    new_data = df1.merge(df2,left_on = 'id',right_on ='match_id')
    df = new_data[new_data['city']==n]
    print(df.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False).head(4))
```

```python
gnd('Mumbai')
```

```python
gnd('Kolkata')
```
